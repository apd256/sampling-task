## Library imports

```python
import numpy as np
import math

import matplotlib.pyplot as plt
from skimage.transform import resize
import cv2

import os
```

## Define functions

### Downsampling

```python
# Image downsampling function
def img_downsampling(desired_size, arr, method="mean"):
    
    # Parameter guard
    if len(arr) <= desired_size or len(arr[0]) <= desired_size:
        raise ValueError("Desired size must not be greater or equal than the size of array!")
    elif len(arr) % 2 != desired_size % 2:
        raise ValueError("Type of number (odd or even) for an array and desired size is not equal!")
    elif len(arr) != len(arr[0]):
        raise ValueError("Must be an equal size or an ordo type!")
    
    result = np.zeros((desired_size, desired_size))

    # Check if the size of an array and its desired size is odd or even
    if desired_size % 2 == 0 and len(arr) % 2 == 0:
        # Begin downsampling for even number size
        for i in range(desired_size):
            for j in range(desired_size):
                
                # block = arr[i * desired_size:(i + 1) * (desired_size), j * desired_size:(j + 1) * (desired_size)]
                block = arr[i*2:(i+1)*2, j*2:(j+1)*2]
                
                if method == 'max':
                    result[i, j] = np.max(block)
                elif method == 'median':
                    result[i, j] = np.median(block)
                elif method == 'mean':
                    result[i, j] = np.mean(block)
    else:
        indices = [(0,2), (2,4), (4,5)]

        # Begin downsampling for odd number size
        for i in range(desired_size):
            for j in range(desired_size):
                r_start, r_end = indices[i]
                c_start, c_end = indices[j]
                
                block = arr[r_start:r_end, c_start:c_end]
                
                if method == 'max':
                    result[i, j] = np.max(block)
                elif method == 'median':
                    result[i, j] = np.median(block)
                elif method == 'mean':
                    result[i, j] = np.mean(block)
                
    return result
```

### Upsampling

```python
# Image upsampling function
def img_upsampling(desired_size, arr, method="nearest"):

    # Parameter guard
    if len(arr) >= desired_size or len(arr[0]) >= desired_size:
        raise ValueError("Desired size must not be less or equal than the size of array!")
    elif len(arr) % 2 != desired_size % 2:
        raise ValueError("Type of number (odd or even) for an array and desired size is not equal!")
    elif len(arr) != len(arr[0]):
        raise ValueError("Must be an equal size or an ordo type!")

    # Check if the size of an array and its desired size is odd or even
    if desired_size % 2 == 0 and len(arr) % 2 == 0:
        if method == 'nearest':
            # return np.repeat(np.repeat(arr, math.sqrt(desired_size), axis=0), math.sqrt(desired_size), axis=1)
            return np.repeat(np.repeat(arr, math.log2(desired_size), axis=0), math.log2(desired_size), axis=1)
        elif method == 'bilinear':
            return resize(arr, (desired_size, desired_size), order=1, preserve_range=True, anti_aliasing=False)
        elif method == 'bicubic':
            return resize(arr, (desired_size, desired_size), order=3, preserve_range=True)
    else:
        if method == 'nearest':
            scale = desired_size / len(arr)
            result = np.zeros((desired_size, desired_size))
            for i in range(desired_size):
                for j in range(desired_size):
                    src_i = int(i / scale)
                    src_j = int(j / scale)
                    result[i, j] = arr[src_i, src_j]
            return result
        elif method == 'bilinear':
            return resize(arr, (desired_size, desired_size), order=1, preserve_range=True, anti_aliasing=False)
        elif method == 'bicubic':
            return resize(arr, (desired_size, desired_size), order=3, preserve_range=True)
```
